**1) 시리즈(Series) :**  
1차원 배열의 값(values)에 각 값에 대응되는 인덱스(index)를 부여할 수 있는 구조

```python
import pandas as pd

sr = pd.Series([17000, 18000, 1000, 5000], index = ["피자", "치킨", "콜라", "맥주"])
print('시리즈 출력 :')
print('-'*15)
print(sr)
print('\n')

#값과 인덱스 출력
print('시리즈의 값 : {}'.format(sr.values))
print('시리즈의 인덱스 : {}'.format(sr.index))
```

**2) 데이터프레임(DataFrame) :**

2차원 리스트를 매개변수로 전달 (행과열)

    행방향 인덱스(index)와 열방향 인덱스(column)가 존재

    열(columns), 인덱스(index), 값(values)으로 구성



```python
values = [[1,2,3], [4,5,6], [7,8,9]]
index = ['one', 'two', 'three']
columns = ['A', 'B', 'C']

df = pd.DataFrame(values, index = index, columns = columns)

print('데이터프레임 출력 :')
print('-'*18)
print(df)
print('\n')

#인덱스, 값, 열 각각 출력
print('데이터프레임의 인덱스 : {}'.format(df.index))
print('데이터프레임의 열이름 : {}'.format(df.columns))

#print('데이터프레임의 값 : {} '.format(df.values))

print('데이터프레임의 값 :')
print('-'*18)
print(df.values)
```

**3) 데이터프레임의 생성 :**  
리스트(List), 시리즈(Series), 딕셔너리(dict), Numpy의 ndarrays, 또 다른 데이터프레임으로부터 생성 가능


```python
# 데이터 프레임 생성 : 리스트 (이중리스트)

data = [
    ['1000', 'Steve', 90.72],
    ['1001', 'James', 78.09],
    ['1002', 'Doyeon', 98.43],
    ['1003', 'Jane', 64.19],
    ['1004', 'Pilwoong', 81.30],
    ['1005', 'Tony', 99.14],
]

df = pd.DataFrame(data)
print(df)

# 생성된 데이터프레임에 열(columns) 지정
print('\n')
df2 = pd.DataFrame(data, columns=['학번', '이름', '점수'])
print(df2)

# 행 지정 (추가)
print('\n')
df3 = pd.DataFrame(data, index = ['1번', '2번', '3번', '4번', '5번', '6번'])
print(df3)
```

```python
# 데이처프레임 생성 : 딕셔너리

data = {
    '학번' : ['1000', '1001', '1002', '1003', '1004', '1005'],
    '이름' : ['Steve', 'James', 'Doyeon', 'Jane', 'Pilwoong', 'Tony'],
    '점수' : [90.72, 78.09, 98.43, 64.19, 81.30, 99.14]
}

df = pd.DataFrame(data)
print(df)
```

**2. 넘파이(Numpy)**

**1) np.array() : Numpy의 핵심**

    리스트, 튜플, 배열로 부터 ndarray를 생성

```python
import numpy as np

# 1차원 배열
vec = np.array([1,2,3,4,5])
print(vec)

# 2차원 배열
print('\n')

mat = np.array([[10, 20, 30],[60, 70,80]])
print(mat)

# 두 배열 타입 확인 : type()
print('\n')
print('vec 타입 :', type(vec))
print('mat 타입 :', type(mat))

"""
타입 : numpy.ndarray
Numpy 배열 : 축의 개수(ndim)와 크기(shape)라는 개념 존재
배열의 크기를 정확히 숙지하는 것은 딥 러닝에서 매우 중요
"""
print('\n')

#축 개수 및 크기 출력
print('vec 축 개수 :', vec.ndim) # 축의 개수 : .ndim
print('vec 크기(shape) :',vec.shape) # 크기 : .shape
print('\n')
print('mat 축 개수 :',mat.ndim)
print('mat 크기(shape) :',mat.shape)
```

**2) ndarray의 초기화**

```python
from shutil import SameFileError
#모든 값이 0인 2X3 배열 생성 : np.zeros((행,열))
zero_mat = np.zeros((2,3))
print(zero_mat)

print('\n')
# 모든 값이 1인 2x3 배열 생성 : np.ones((행,열))
one_mat = np.ones((2,3))
print(one_mat)

print('\n')
# 모든 값이 특정 상수인 배열 생성 : np.full((행,열), 상수)
same_value_mat = np.full((2,2), 7)
print(same_value_mat)

print('\n')
# 대각선 값이 1이고 나머지 값이 0인 2차원 배열을 생성 : np.eye(행열 수)
eye_mat = np.eye(3)
print(eye_mat)

print('\n')
# 임의의 값으로 채워진 배열 생성 : np.random.random()
random_mat = np.random.random((2,2))
print(random_mat)
```

**3) np.arange()**

```python
# 0부터 9까지
range_vcec = np.arange(10)
print(range_vcec)

print('\n')
#1부터 9까지 +2씩 적용
n = 2
range_step_vec = np.arange(1,10,n)
print(range_step_vec)
```

**4) np.reshape()**

내부 데이터는 변경하지 않으면서 배열의 구조를 바꿈

```python
reshape_mat = np.array(np.arange(30)).reshape((5,6))
print(reshape_mat)
```

**5) Numpy 슬라이싱**

슬라이싱 기능을 사용해 특정 행이나 열들의 원소들을 접근

```python
mat = np.array([[1, 2, 3], [4, 5, 6]])
print(mat)

print('\n')
#첫번째 행 출력
slicing_mat = mat[0, :]
print(slicing_mat)

print('\n')
#두번째 열 출력
slicing_mat2 = mat[:, 1]
print(slicing_mat2)
```

**6) Numpy 정수 인덱싱(integer indexing)**

연속적이지 않은 원소로 배열을 만들 경우에는 슬라이싱으로는 만들 수 X -> 인덱싱을 사용하여 배열을 구성

```python
mat = np.array([[1,2], [4,5], [7,8]])
print(mat)

#특정 위치 원소 가져오기
# 1행 0열의 원소 => 0부터 카운트하므로 두번째 행 첫번째 열의 원소.
print('\n',mat[1,0])

# mat[[2행, 1행],[0열, 1열]]
# 각 행과 열의 쌍을 매칭하면 2행 0열, 1행 1열의 두 개의 원소.
indexing_mat = mat[[2,1],[0,1]]
print('\n',indexing_mat)
```

**7) Numpy 연산**


```python
import numpy as np

x = np.array([1,2,3])
y = np.array([4,5,6])

# 덧셈
result = np.add(x,y)
print(result)
# 뺄셈
result = np.subtract(x,y)
print(result)
# 곱셈
result = np.multiply(result,x) # 요소별 곱
print(result)
# 나눗셈
result = np.divide(result,x)
print(result)

# 벡터와 행렬곱 : dot()
mat1 = np.array([[1,2],[3,4]])
mat2 = np.array([[5,6],[7,8]])
mat3 = np.dot(mat1, mat2)
print(mat3)
```

데이터를 차트(chart)나 플롯(plot)으로 시각화하는 패키지

**1) 라인 플롯 그리기**

plot() : 라인 플롯을 그리는 기능

show() : 시각화 #주피터 노트북은 사용안해도 그래프 렌더링

```python
import matplotlib.pyplot as plt
# 라인 플롯 그리기
plt.title('test')
plt.plot([1, 2, 3, 4], [2, 4, 8, 6])

# 축 레이블 삽입
plt.xlabel('hours')
plt.ylabel('score')
plt.show()
```

**3) 라인 추가와 범례 삽입하기**

legend() : 라인 플롯을 동시에 사용할 경우에는 각 선이 어떤 데이터를 나타내는지를 보여주기 위해 사용 (범례)

```python
plt.title('students')
plt.plot([1,2,3,4],[2,4,8,6])
plt.plot([1.5, 2.5, 3.5, 4.5], [3,5,8,10]) #라인 새로 추가
plt.xlabel('hours')
plt.ylabel('score')
plt.legend(['A student', 'B student']) #범례 삽입
plt.show()
```
