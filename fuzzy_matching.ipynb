# 전체 데이터셋 평가

```python
# 70이상

import fitz  # PyMuPDF
import json
from fuzzywuzzy import fuzz

# PDF 파일 열기 및 텍스트 추출
def extract_text_from_pdf(pdf_path):
    pdf_document = fitz.open(pdf_path)
    pdf_text = ""
    for page_num in range(pdf_document.page_count):
        page = pdf_document.load_page(page_num)
        pdf_text += page.get_text()
    return pdf_text

# 텍스트 청크 나누기
def chunk_text(text, chunk_size=1000):
    return [text[i:i+chunk_size] for i in range(0, len(text), chunk_size)]

# 유사도 계산
def calculate_similarity(reference, hypothesis):
    return fuzz.partial_ratio(reference, hypothesis)

# JSON 파일 읽기
def load_json(json_path):
    with open(json_path, 'r', encoding='utf-8') as file:
        return json.load(file)

# PDF에서 텍스트 추출
pdf_text = extract_text_from_pdf("D:/[24]ICT_Practice/practice_file/(2024)포트미스_가이드북.pdf")
# 텍스트를 청크 단위로 나누기
chunks = chunk_text(pdf_text)
# JSON 데이터 로드
qa_data = load_json("D:/[24]ICT_Practice/fineturing_ex/dataset/(2024)gpt4o_fineturning_data.json")

# 유사도 계산 및 평가
results = []
correct_count = 0
total_similarity = 0

for entry in qa_data:
    question = entry["QUESTION"]
    answer = entry["ANSWER"]
    
    max_similarity = 0
    best_chunk_num = None
    
    # 모든 청크에서 정답과 유사한 부분 찾기
    for chunk_num, chunk in enumerate(chunks):
        similarity = calculate_similarity(chunk, answer)
        if similarity > max_similarity:
            max_similarity = similarity
            best_chunk_num = chunk_num + 1

    results.append({
        "question": question,
        "answer": answer,
        "similarity": max_similarity,
        "chunk_number": best_chunk_num
    })
    
    # 유사도 기준 이상인 경우를 올바르게 추출된 것으로 간주
    if max_similarity >= 70:
        correct_count += 1
    
    total_similarity += max_similarity

# 평가 메트릭 계산
total_questions = len(qa_data)
accuracy = correct_count / total_questions
mean_similarity = total_similarity / total_questions

# 평가 결과 출력
print(f"Total Questions: {total_questions}")
print(f"Correctly Extracted: {correct_count}")
print(f"Accuracy: {accuracy * 100:.2f}%")
print(f"Mean Similarity: {mean_similarity:.2f}")
```

```python
# 상세 결과 출력
for result in results:
    print(f"Question: {result['question']}")
    print(f"Answer: {result['answer']}")
    print(f"Similarity: {result['similarity']}")
    if result["chunk_number"]:
        print(f"Found in Chunk: {result['chunk_number']}\n")
    else:
        print("Not Found in any Chunk\n")
```
